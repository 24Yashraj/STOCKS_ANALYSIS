```python
dfs = [Apple, Nvdia, Fb, Tesla, Twitter, Google]
```

```python
for df in dfs:
        df['Mcn50'] = df.Close.rolling(50).mean()
        df['Mcn200'] = df.Close.rolling(200).mean()
```

```python
Apple.head(51)
```

```python
for df in dfs:
        df['Previous Day close Price'] = df.Close.shift(1)
```

```python
Apple.head()
```

```python
for df in dfs:
        df['Change in price'] = df['Close'] - df['Previous Day close Price']
```

```python
Apple.head()
```

```python
for df in dfs:
        df['Percent Change in price'] = df.Close.pct_change()
```

```python
for df in dfs:
        df['Pervious Day Volume'] = df.Volume.shift(1)
```

```python
for df in dfs:
       df['Change in Volume'] = df['Volume']  -  df['Pervious Day Volume']
```
